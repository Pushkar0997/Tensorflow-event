## What are AI, ML, and Deep Learning?

Let’s start with the big umbrella terms you hear everywhere.

- **Artificial Intelligence (AI)**: Any technique that enables computers to mimic human intelligence — reasoning, perception, decision-making, problem-solving.
- **Machine Learning (ML)**: A **subset** of AI where systems learn patterns from data instead of being fully hand-coded with if-else rules.
- **Deep Learning (DL)**: A **subset** of ML that uses deep neural networks (many layers of artificial “neurons”). It is especially powerful for images, speech, and natural language.

A common way to visualize this is:

![AI vs ML vs DL](https://cdn.analyticsvidhya.com/wp-content/uploads/2024/10/76182pic1.webp)

You can think of it like this:

- AI is the overall goal: making machines “smart”.
- ML is one way to achieve that, by learning from data.
- DL is a specific ML technique that has driven recent breakthroughs (vision, speech, large language models, etc.).

## The Machine Learning Workflow

Almost every ML project, from a college project to an industry system, follows a similar high-level workflow.

Typical stages:

1. **Problem definition** – What are we trying to predict or decide?
2. **Data collection** – Get relevant examples (rows) and information (columns).
3. **Data understanding & exploration** – Inspect, visualize, and understand patterns and issues.
4. **Data cleaning & preprocessing** – Handle missing values, inconsistent types, outliers.
5. **Feature engineering** – Transform raw data into useful inputs for the model.
6. **Model selection & training** – Choose an algorithm and learn from data.
7. **Evaluation** – Check how well the model performs on unseen data.
8. **Deployment & monitoring** – Put the model into real use and keep improving it.

## What kinds of problems can ML solve?

Once we have data in a table, we can use ML to solve several types of problems.

Some important ones:

- **Classification**: Predict a category.
  - Example: pass vs fail, spam vs not spam, disease vs no disease.

  - Example: grouping customers into segments for marketing.

Visually, you might see diagrams that show:

- Regression as a line going through points.
- Classification as colored regions separating classes.

## Why is data so important?

Traditional programming:

- You write **rules** (`if`, `else`, `for`, `while`) + input → the computer produces the output.

Machine learning:

- You give the computer **data + examples of correct answers**, and it **learns the rules** by itself.

This is why:

- Good, clean, representative **data** matters more than fancy algorithms.
- Deep learning, in particular, shines when you have **a lot of labeled data**.

## Our first tiny dataset (students and marks)

To keep things simple and fully controllable, we will **create** a tiny dataset directly in code.

Imagine we surveyed some students and recorded:

- `hours_studied` – how many hours they studied for an exam.
- `previous_score` – their score in a previous test.
- `attendance` – attendance percentage.
- `final_score` – their final exam score.
- A simple label `passed` (1) or `failed` (0) based on the final score.

We will start by creating a small table using pandas so you can see:

- Rows = students.
- Columns = features and target.


```python
import pandas as pd

# Create a tiny synthetic dataset of students
data = {
    "student_id": list(range(1, 11)),
    "hours_studied": [1, 2, 3, 4, 5, 2, 6, 7, 3, 4],
    "previous_score": [40, 45, 50, 55, 60, 42, 70, 80, 48, 53],
    "attendance": [60, 65, 70, 75, 80, 62, 85, 90, 68, 72],
    "final_score": [45, 48, 52, 58, 65, 47, 75, 82, 50, 56],
}

df = pd.DataFrame(data)

# Create a simple pass/fail label based on final_score
df["passed"] = (df["final_score"] >= 50).astype(int)

df.head()
```

### Understanding this table

From the output above, notice:

- Each **row** represents one student.
- Each **column** contains a piece of information:
  - `hours_studied`, `previous_score`, `attendance`, `final_score` are numeric **features**.
  - `passed` is a **label** for a simple classification problem (1 = passed, 0 = failed).

In a real-world project:

- This data might come from a CSV exported from your college database.
- You would often have **more rows**, more **features**, and sometimes **missing values** or inconsistencies.

For now, we keep it small so we can focus on the idea:

> Can a model learn a pattern: students who study more / have better history / higher attendance are more likely to pass?


```python
import numpy as np

# Features (X) and labels (y)
feature_cols = ["hours_studied", "previous_score", "attendance"]
X = df[feature_cols].values.astype("float32")
y = df["passed"].values.astype("float32")

X, y
```

## Our first prediction with TensorFlow/Keras

Now we will let a **small neural network** learn from this data.

Even if you do not understand the details yet, focus on these ideas:

- We give the model **inputs** (hours studied, previous score, attendance).
- We give it **targets** (pass or fail).
- The model adjusts its **internal parameters** (weights and biases) to reduce error on the training data.

We will use:

- **TensorFlow** as the underlying deep learning framework.
- The **Keras API** to define and train our model with just a few lines of code.

Next, we will build a tiny network using the `Sequential` API.


```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Simple Sequential model for binary classification
model = keras.Sequential(
    [
        layers.Input(shape=(len(feature_cols),)),
        layers.Dense(8, activation="relu"),
        layers.Dense(4, activation="relu"),
        layers.Dense(1, activation="sigmoid"),  # output between 0 and 1 for "pass probability"
    ]
)

model.summary()
```

```python
# Compile the model: define loss, optimizer, and metric
model.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=["accuracy"],
)

# Train for a few epochs (tiny dataset, so this will be very fast)
history = model.fit(
    X,
    y,
    epochs=30,
    batch_size=4,
    verbose=0,  # change to 1 if you want to show logs
)

# Evaluate on the same small dataset (just for demo)
loss, acc = model.evaluate(X, y, verbose=0)
print(f"Training accuracy on this tiny dataset: {acc:.2f}")
```

```python
# Let's make predictions for a few hypothetical students
test_students = np.array(
    [
        [2, 40, 60],  # low hours, low previous score, low attendance
        [6, 70, 85],  # high hours, high previous score, high attendance
        [3, 50, 70],  # medium everything
    ],
    dtype="float32",
)

pred_probs = model.predict(test_students)
pred_labels = (pred_probs >= 0.5).astype(int)

for i, (features, prob, label) in enumerate(zip(test_students, pred_probs, pred_labels)):
    print(
        f"Student {i+1} -> hours_studied={features[0]}, previous_score={features[1]}, attendance={features[2]} "
        f"=> pass_probability={prob[0]:.2f}, predicted_label={int(label[0])}"
    )
```
