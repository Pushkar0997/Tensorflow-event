```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils import resample
```

```python
# Load dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df.head()
```

```python
# ---------------- EDA ----------------
print(df.head())
print(df.info())
print(df.describe())
```

```python
# ---------------- Cleaning ----------------
df = df.drop_duplicates()
```

```python
# ---------------- Transformation ----------------
scaler = StandardScaler()
X = scaler.fit_transform(df.drop('target', axis=1))
y = df['target']
```

```python
# ---------------- Splitting ----------------
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# ---------------- Balancing (example) ----------------
df_train = pd.DataFrame(X_train)
df_train['target'] = y_train.values

majority = df_train[df_train['target'] == 0]
minority = df_train[df_train['target'] == 1]

minority_upsampled = resample(minority, replace=True, n_samples=len(majority))

df_balanced = pd.concat([majority, minority_upsampled])

print("Balanced class count:")
print(df_balanced['target'].value_counts())
```

##  Classification in Machine Learning

### What is Classification?
Classification is a type of ML where the model assigns data into categories (classes).

Example:
- Spam or Not Spam  
- Cat or Dog  
- Pass or Fail  

### How Classification Works

1. Collect Data  
   - Input data + labels (correct answers)

2. Train Model  
   - Model learns patterns from labeled data

3. Test Model  
   - Check how well it predicts on new data

4. Predict  
   - Model classifies new unseen input


### Types of Classification
- Binary → 2 classes (Yes/No)  
- Multi-class → More than 2 classes  
- Multi-label → Multiple labels at once  

#### Common Algorithms
- Logistic Regression

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 1. Load dataset
data = load_iris()
X = data.data
y = data.target

# 2. Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 3. Train model
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# 4. Predict
y_pred = model.predict(X_test)

# 5. Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# 6. Try your own prediction
sample = [[5.1, 3.5, 1.4, 0.2]]
print("Predicted class:", model.predict(sample))
```

# Linear Regression

Linear Regression is used to predict a continuous value based on input features.

It tries to fit a straight line that best represents the relationship between input (X) and output (y).


## Idea
The model learns how changes in input affect the output.

Example:
- More area → higher house price  
- More experience → higher salary  


## Steps
1. Take input features (X) and target (y)  
2. Fit a best line to the data  
3. Use the line to predict new values  


Link : https://observablehq.com/@yizhe-ang/interactive-visualization-of-linear-regression

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load dataset
data = fetch_california_housing()
X = data.data
y = data.target

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
pred = model.predict(X_test)

# Show sample output
print("Sample predictions:", pred[:5])
```

# Logistic Regression

Logistic Regression is used for classification problems.

It predicts the probability that an input belongs to a certain class.

## Idea
Instead of predicting a number, it predicts a class.

Example:
- Email → Spam or Not Spam  
- Student → Pass or Fail  

## How it Works
1. Takes input features (X)  
2. Calculates probability (0 to 1)  
3. Converts probability into a class  


Link: https://ai.williamtheisen.com/logistic-regression.html

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.datasets import fetch_openml
from sklearn.linear_model import LogisticRegression
import numpy as np

# Load dataset
data = fetch_openml('heart-statlog', version=1, as_frame=False)

# Select features: Age, BP, Cholesterol
X = data.data[:, [0, 3, 4]]

# Convert target to 0/1
y = np.where(data.target == 'present', 1, 0)

# Train model
model = LogisticRegression(max_iter=1000)
model.fit(X, y)

# Test with sample input
patient = [[45.0, 120.0, 180.0]]

prob = model.predict_proba(patient)[0][1]

print("Probability of heart disease:", prob)

# Simple prediction
if prob > 0.5:
    print("High risk")
else:
    print("Low risk")
```

# K-Means Clustering

K-Means is an unsupervised learning algorithm used to group data into clusters.

It does not use labeled data.

## Idea
The model groups similar data points together.

Example:
- Customer segmentation  
- Grouping similar items  


## How it Works
1. Choose number of clusters (K)  
2. Randomly assign cluster centers  
3. Assign points to nearest cluster  
4. Update cluster centers  
5. Repeat until stable  


## Output
The model assigns each data point to a cluster.

Link for Visualization: https://www.naftaliharris.com/blog/visualizing-k-means-clustering/

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Load dataset
data = load_iris()
X = data.data

# Apply KMeans
model = KMeans(n_clusters=3)
model.fit(X)

# Get cluster labels
labels = model.labels_

print("Cluster labels:", labels[:10])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_sample_image

# Load image (DON'T scale here)
image = load_sample_image("china.jpg")

w, h, d = image.shape

# Flatten pixels
pixels = image.reshape(-1, d)

# Apply KMeans
model = KMeans(n_clusters=16, random_state=42, n_init=10)
labels = model.fit_predict(pixels)

# Reconstruct image
compressed = model.cluster_centers_[labels].reshape(w, h, d)

# Convert to uint8 (VERY IMPORTANT)
compressed = np.clip(compressed, 0, 255).astype('uint8')

# Show images
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].imshow(image)
ax[0].set_title("Original")
ax[0].axis("off")

ax[1].imshow(compressed)
ax[1].set_title("Compressed (8 colors)")
ax[1].axis("off")

plt.show()
```

## From Machine Learning to Deep Learning

So far, we have:

- Worked with **tabular data**.
- Built a simple baseline model (logistic regression) for classification.

Deep learning is:

- A **subset of machine learning**.
- Uses multi-layer **neural networks** to learn complex patterns.
- Especially strong on **images, text, audio, and large datasets**.

Key differences (very high level):

- Classical ML:
  - Often needs a lot of **manual feature engineering**.
  - Works very well on smaller, structured datasets.
- Deep Learning:
  - Learns **features automatically** from raw data (pixels, words, signals).
  - Needs **more data** and **more compute**, but can reach very high performance.

## Why Machine Learning Was Not Enough

Traditional ML has some limitations:

- Requires manual feature selection  
- Struggles with complex data (images, audio, text)  
- Performance depends heavily on human effort  
- Not fully end-to-end


# Deep Learning Workflow

## Idea
Deep Learning models learn patterns directly from data and make predictions.

---

## Steps

### 1. Data Collection
Get data (images, text, numbers)

### 2. Data Preprocessing
Clean and prepare data:
- Normalize values
- Convert into proper format

### 3. Build Model
Create a neural network:
- Input layer
- Hidden layers
- Output layer

### 4. Training
Model learns by adjusting weights using data

### 5. Evaluation
Check how well the model performs on new data

### 6. Prediction
Use trained model to make predictions

---

## Flow
Data → Model → Training → Evaluation → Prediction



```python
import tensorflow as tf
import numpy as np

# 1. Data
X = np.array([[0], [1], [2], [3], [4]], dtype=float)
y = np.array([0, 2, 4, 6, 8], dtype=float)

# 2. Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1])
])

# 3. Compile
model.compile(optimizer='sgd', loss='mean_squared_error')

# 4. Train
model.fit(X, y, epochs=100, verbose=0)

# 5. Predict
print("Prediction for 5:", model.predict(np.array([[5.0]]), verbose=0))
```

# Perceptron

## What is a Perceptron?
A Perceptron is the simplest type of neural network used for binary classification.

It takes input features and produces a single output (0 or 1).


## How it Works

1. Takes input values (X)  
2. Multiplies with weights (W)  
3. Adds bias (b)  
4. Passes through an activation function  


## Formula
Output = activation(W·X + b)


## Idea
It tries to find a line that separates two classes.


## Limitation
- Works only for linear data  
- Cannot solve non-linear problems

```python
import tensorflow as tf
import numpy as np

# Sample data (AND logic)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=float)

y = np.array([0, 0, 0, 1], dtype=float)

# Build model (single neuron = perceptron)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compile
model.compile(optimizer='sgd', loss='binary_crossentropy')

# Train
model.fit(X, y, epochs=100, verbose=0)

# Predict
pred = model.predict(X)

print("Predictions:")
print(pred)
```

# Artificial Neural Network (ANN)

## What is ANN?
ANN is a basic deep learning model made of layers of neurons.

It is used for learning patterns from data.


## Structure

### 1. Input Layer
Takes input data

### 2. Hidden Layers
Learns patterns by adjusting weights

### 3. Output Layer
Gives final prediction


![Image](https://cdn.analyticsvidhya.com/wp-content/uploads/2025/05/4-4.webp)

## How it Works
1. Input goes into the network  
2. Each layer processes the data  
3. Model learns by adjusting weights  
4. Final output is predicted  



## Example
Input → features → hidden layers → output (class/value)


![Image](https://cdn.analyticsvidhya.com/wp-content/uploads/2025/05/3-6.webp)



## Limitation
ANN works well for simple data but is not ideal for images (CNN is better).





```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Load dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# Preprocess
X_train = X_train / 255.0
X_test = X_test / 255.0

# Flatten (since ANN needs 1D input)
X_train = X_train.reshape(-1, 28*28)
X_test = X_test.reshape(-1, 28*28)

# Build ANN model
model = models.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train
model.fit(X_train, y_train, epochs=3)

# Evaluate
model.evaluate(X_test, y_test)
```

```python
# Install yfinance to pull real market data
!pip install yfinance -q
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler

# 1. Download REAL Stock Data from Yahoo Finance
ticker_symbol = "AAPL" # Apple Inc.
print(f"Downloading real market data for {ticker_symbol}...")
data = yf.download(ticker_symbol, start="2022-01-01", end="2024-01-01", progress=False)


real_prices = data['Close'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_prices = scaler.fit_transform(real_prices)

# 2. Prepare Data: Use a 10-day window to predict the 11th day
window_size = 10
X, y = [], []
for i in range(len(scaled_prices) - window_size):
    X.append(scaled_prices[i : i + window_size, 0])
    y.append(scaled_prices[i + window_size, 0])

X = np.array(X)
y = np.array(y)

# 3. Build the AI Trader (ANN)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu', input_shape=(window_size,)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1) # Output is a single predicted price
])

model.compile(optimizer='adam', loss='mse')

# 4. Train the AI on the real data
print("Training AI on actual Wall Street history... please wait.")
model.fit(X, y, epochs=50, batch_size=16, verbose=0)
print("Training Complete!\n")

# Predict the history to see how well it tracked reality
predicted_scaled = model.predict(X, verbose=0)
predicted_prices = scaler.inverse_transform(predicted_scaled) # Convert back to real dollars
actual_prices = scaler.inverse_transform(y.reshape(-1, 1))

# 5. Visualize the AI vs Reality
plt.figure(figsize=(12, 5))
plt.plot(actual_prices, label=f"Real {ticker_symbol} Price", color='black', linewidth=2)
plt.plot(predicted_prices, label="AI Prediction", color='red', linestyle='dashed', linewidth=2)

plt.title(f"Real-World AI Stock Tracker: {ticker_symbol}")
plt.xlabel("Days")
plt.ylabel("Stock Price ($)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

# Convolutional Neural Network (CNN)

## What is CNN?
CNN is a deep learning model designed for image data.

Unlike ANN, it does not flatten the image at the start.  
It keeps the spatial structure of the image.


## Why CNN over ANN?
- ANN loses image structure (by flattening)  
- CNN keeps spatial information  
- Better performance on images  


## How it Works

### 1. Convolution Layer
Extracts features like edges, shapes

### 2. Pooling Layer
Reduces size and keeps important features

### 3. Flatten Layer
Converts data into 1D

### 4. Dense Layer
Makes final prediction


## Flow
Image → Convolution → Pooling → Flatten → Dense → Output


## Example
Image → classify digit (0–9)


## Conclusion
CNN is powerful for image recognition and is widely used in computer vision.


```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Load dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# Preprocess
X_train = X_train / 255.0
X_test = X_test / 255.0

# Add channel dimension
X_train = X_train[..., tf.newaxis]
X_test = X_test[..., tf.newaxis]

# Build CNN model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train
model.fit(X_train, y_train, epochs=3)

# Evaluate
model.evaluate(X_test, y_test)
```

# Recurrent Neural Network (RNN)

## What is RNN?
RNN is a deep learning model used for sequential data.

It remembers previous information to make better predictions.

## Use Cases
- Text prediction  
- Language translation  
- Time series (stock, weather)  

## Idea
Unlike ANN/CNN, RNN uses past data (memory).

Example:
Previous words → predict next word

## How it Works
1. Takes input one step at a time  
2. Keeps memory of previous inputs  
3. Uses that memory for prediction  

## Flow
Sequence → RNN → Output

## Conclusion
RNN is useful when order and sequence matter.

```python
import tensorflow as tf
import numpy as np

# Set a random seed so the results are consistent every time you run it!
tf.random.set_seed(42)
np.random.seed(42)

# 1. A tiny dataset of movie reviews
reviews = [
    "I loved this movie it was fantastic",
    "Terrible film, a complete waste of time",
    "The acting was brilliant and the story was great",
    "Awful, boring, and completely unwatchable",
    "Best movie of the year",
    "I hated every minute of it",
    "A wonderful experience",
    "Do not watch this garbage"
]
labels = np.array([1, 0, 1, 0, 1, 0, 1, 0]) # 1 = Positive, 0 = Negative

# 2. Text Vectorization (Convert words into numbers)
vectorizer = tf.keras.layers.TextVectorization(max_tokens=50, output_sequence_length=10)
vectorizer.adapt(reviews)
X = vectorizer(reviews)

# 3. Build the RNN Model
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=50, output_dim=8),
    tf.keras.layers.SimpleRNN(16),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# 4. Compile and Train
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# FIX: Increased epochs from 50 to 200 so the model fully learns the words
model.fit(X, labels, epochs=200, verbose=0)

# 5. Predict on NEW custom text!
test_reviews = [
    "I loved it",
    "It was terrible and boring"
]

X_test = vectorizer(test_reviews)
predictions = model.predict(X_test, verbose=0)

# 6. Show Results
for text, pred in zip(test_reviews, predictions):
    sentiment = "Positive" if pred > 0.5 else "Negative"
    print(f"Review: '{text}' \nScore: {pred[0]:.2f} -> {sentiment}\n")
```
